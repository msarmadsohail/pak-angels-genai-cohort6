Code for Entire ChatBot

```python
# üìå Import required modules
import getpass
from IPython.display import display, Markdown
from openai import OpenAI

# üìå Ask for Groq API Key
# groq_api_key = getpass.getpass("üîë Enter your Groq API Key: ")
groq_api_key = API_KEY
# üìå Create Groq-compatible client
client = OpenAI(
    api_key=groq_api_key,
    base_url="https://api.groq.com/openai/v1",
)

# üìå Initialize chat history
chat_history = [
    {"role": "system", "content": "You are a helpful assistant powered by GroqCloud."}
]

# üìå Chat function using Groq-compatible client
def chat_with_groq(user_input, model="llama3-8b-8192"):
    chat_history.append({"role": "user", "content": user_input})

    try:
        response = client.chat.completions.create(
            model=model,
            messages=chat_history,
            temperature=0.7,
        )
        reply = response.choices[0].message.content
        chat_history.append({"role": "assistant", "content": reply})
        return reply

    except Exception as e:
        return f"‚ö†Ô∏è Error: {str(e)}"

# üìå Main Chat Loop
print("üí¨ Start chatting with GroqBot! (Type 'exit' to quit)\n")
while True:
    user_prompt = input("You: ")
    if user_prompt.strip().lower() in ["exit", "quit"]:
        print("üëã Goodbye!")
        break
    reply = chat_with_groq(user_prompt)
    display(Markdown(f"**ü§ñ GroqBot:** {reply}"))
```

# English to Urdu Translator

```python
# ‚úÖ Install the latest OpenAI SDK and Gradio
!pip install --upgrade openai gradio --quiet
```

```python
from google.colab import userdata
API_KEY=userdata.get('ChatBot_using_GroqAPI')
```

```python
# üìå Set your Groq API Key
GROQ_API_KEY = API_KEY  # Replace with your key

# ‚úÖ Setup OpenAI client to use Groq API
import openai
from openai import OpenAI

client = OpenAI(
    api_key=GROQ_API_KEY,
    base_url="https://api.groq.com/openai/v1"  # Required for Groq
)

# üåç Supported target languages
LANGUAGES = {
    "Spanish": "es",
    "French": "fr",
    "German": "de",
    "Chinese": "zh",
    "Urdu": "ur",
    "Arabic": "ar",
    "Hindi": "hi",
    "Russian": "ru",
    "Japanese": "ja",
    "Italian": "it",
    "Korean": "ko"
}

# üîÑ Translation function
def translate_text(text, target_lang):
    if not text:
        return "Please enter some text."

    prompt = f"""Translate the following English text to {target_lang}:\n\n\"{text}\"\n\nTranslation:"""

    try:
        response = client.chat.completions.create(
            model="llama3-70b-8192",  # ‚úÖ RECOMMENDED GROQ MODEL
            messages=[
                {"role": "user", "content": prompt}
            ],
            temperature=0.3,
            max_tokens=200
        )
        translated = response.choices[0].message.content.strip()
        return translated
    except Exception as e:
        return f"‚ùå Error: {str(e)}"

# üé® Gradio UI
import gradio as gr

with gr.Blocks() as demo:
    gr.Markdown("## üåç English to Language Translator (Powered by Groq API)")
    with gr.Row():
        input_text = gr.Textbox(label="Enter English Text", lines=4, placeholder="Type something in English...")
        output_text = gr.Textbox(label="Translated Text", lines=4)
    lang_dropdown = gr.Dropdown(choices=list(LANGUAGES.keys()), label="Select Language", value="Spanish")
    translate_button = gr.Button("üîÅ Translate")

    translate_button.click(fn=translate_text, inputs=[input_text, lang_dropdown], outputs=output_text)

demo.launch()
```
