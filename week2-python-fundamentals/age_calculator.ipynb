```python
import streamlit as st
from datetime import datetime, date, time
import pytz

def calculate_age(birth_datetime, now_datetime):
    delta = now_datetime - birth_datetime
    total_seconds = delta.total_seconds()

    years = now_datetime.year - birth_datetime.year
    months = now_datetime.month - birth_datetime.month
    days = now_datetime.day - birth_datetime.day
    hours = now_datetime.hour - birth_datetime.hour
    minutes = now_datetime.minute - birth_datetime.minute

    if minutes < 0:
        minutes += 60
        hours -= 1
    if hours < 0:
        hours += 24
        days -= 1
    if days < 0:
        months -= 1
        previous_month = (now_datetime.month - 1) or 12
        previous_year = now_datetime.year if now_datetime.month > 1 else now_datetime.year - 1
        days += (date(previous_year, previous_month + 1, 1) - date(previous_year, previous_month, 1)).days
    if months < 0:
        months += 12
        years -= 1

    return years, months, days, hours, minutes

def main():
    st.set_page_config(page_title="ðŸŒ Accurate Age Calculator", page_icon="ðŸ•°ï¸", layout="centered")
    st.title("ðŸ•°ï¸ Accurate Age Calculator with Time Zone Support")
    st.markdown("Enter your birth date and (optionally) your time and time zone for a precise age.")

    # Birthdate input
    default_birthdate = date(2000, 1, 1)
    birthdate = st.date_input("ðŸ“… Select your birthdate", value=default_birthdate, min_value=date(1900, 1, 1), max_value=date.today())

    # Optional time and timezone input
    with st.expander("ðŸ•‘ Optionally, enter your time of birth"):
        include_time = st.checkbox("Include birth time?")
        if include_time:
            col1, col2 = st.columns(2)
            with col1:
                hour = st.selectbox("Hour", list(range(0, 24)), index=12)
            with col2:
                minute = st.selectbox("Minute", list(range(0, 60)), index=0)
            birth_time = time(hour, minute)

            # Time zone selection
            timezone = st.selectbox("ðŸŒ Select your birth time zone", pytz.all_timezones, index=pytz.all_timezones.index('Asia/Karachi'))
        else:
            birth_time = time(0, 0)
            timezone = 'UTC'

    if st.button("ðŸŽ¯ Calculate Age"):
        # Combine date and time
        birth_dt_naive = datetime.combine(birthdate, birth_time)
        user_tz = pytz.timezone(timezone)
        birth_dt = user_tz.localize(birth_dt_naive)

        # Convert to current local timezone for comparison
        now_local = datetime.now(pytz.utc).astimezone()

        # Calculate
        years, months, days, hours, minutes = calculate_age(birth_dt.astimezone(now_local.tzinfo), now_local)

        age_text = f"ðŸŽ‰ You are {years} years, {months} months, {days} days"
        if include_time:
            age_text += f", {hours} hours, and {minutes} minutes"
        age_text += " old."

        st.success(age_text)

if __name__ == "__main__":
    main()
```
