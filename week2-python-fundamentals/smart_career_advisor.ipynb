# Smart Career Advisor App

Write app.py

```python
# Create app.py
app_code = '''
import streamlit as st

# App Title
st.title("üéì Smart Career Advisor")

# Sidebar Info
st.sidebar.markdown("### Please provide the following details:")

# User Inputs
name = st.sidebar.text_input("Your Name:")
education = st.sidebar.selectbox("Education Level", ["High School", "Bachelor's", "Master's", "PhD"])
interests = st.sidebar.multiselect("Select Your Interests", ["Technology", "Science", "Art", "Business", "Health", "Education", "Law", "Engineering", "Writing"])
future_goal = st.sidebar.text_input("What is your future goal?")
work_style = st.sidebar.selectbox("Preferred Work Style", ["Remote", "On-site", "Hybrid", "Freelance"])

# Rule-Based Logic
def suggest_careers(edu, intrs, goal, work):
    careers = []
    resources = []
    roles = []

    if "Technology" in intrs:
        careers.append("Software Developer")
        roles.append("Frontend Developer, Data Analyst, AI Engineer")
        resources.append("Coursera Python, freeCodeCamp, Harvard CS50")

    if "Business" in intrs:
        careers.append("Business Analyst")
        roles.append("Product Manager, Marketing Analyst, Entrepreneur")
        resources.append("Google Business Courses, MBA Essentials, HubSpot Academy")

    if "Art" in intrs:
        careers.append("Graphic Designer")
        roles.append("UI/UX Designer, Animator, Illustrator")
        resources.append("Canva Design School, Adobe Creative Cloud Tutorials")

    if "Health" in intrs:
        careers.append("Healthcare Professional")
        roles.append("Nurse, Therapist, Lab Technician")
        resources.append("Khan Academy Health, Medical Prep Resources")

    if "Writing" in intrs:
        careers.append("Content Writer")
        roles.append("Copywriter, Blogger, Technical Writer")
        resources.append("Grammarly Blog, Copywriting Courses on Udemy")

    if not careers:
        careers.append("Career Counselor Recommended")
        roles.append("Please talk to an expert based on your unique profile.")
        resources.append("LinkedIn Learning, Coursera, edX")

    return careers, roles, resources

# Show Results
if st.sidebar.button("Get Career Advice"):
    st.subheader(f"üëã Hello, {name}!")

    st.markdown("### üéØ Based on your inputs, here‚Äôs your career guidance:")

    careers, roles, resources = suggest_careers(education, interests, future_goal, work_style)

    with st.expander("üìå Suggested Career Paths"):
        for c in careers:
            st.write(f"- {c}")

    with st.expander("üíº Potential Job Roles"):
        for r in roles:
            st.write(f"- {r}")

    with st.expander("üìö Recommended Learning Resources"):
        for res in resources:
            st.write(f"- {res}")

    with st.expander("üìù Your Profile Summary"):
        st.markdown(f"**Education Level:** {education}")
        st.markdown(f"**Interests:** {', '.join(interests)}")
        st.markdown(f"**Future Goal:** {future_goal}")
        st.markdown(f"**Preferred Work Style:** {work_style}")
'''

with open("app.py", "w") as f:
    f.write(app_code.strip())

print("‚úÖ app.py created.")
```

Write requirements.txt

```python
# Create requirements.txt
with open("requirements.txt", "w") as f:
    f.write("streamlit\n")

print("‚úÖ requirements.txt created.")
```

# Mini Game Arcade

Write app.py

```python
# Create app.py
app_code = '''
import streamlit as st
import random

# Set layout and page config
st.set_page_config(page_title="Mini Game Arcade", layout="wide")

# Inject CSS styling
st.markdown("""
    <style>
    .left-box {
        background-color: #f5f6fa;
        padding: 20px;
        border-radius: 10px;
    }
    .stButton > button {
        width: 100%;
        margin-bottom: 10px;
    }
    </style>
""", unsafe_allow_html=True)

st.markdown("## üéÆ Mini Game Arcade")

# List of games
games = [
    "Rock Paper Scissors", "Guess the Number", "Word Scramble", "Emoji Quiz",
    "Score Tracker", "Dice Roller", "Coin Toss", "Math Quiz",
    "Hangman", "Color Guess"
]

# Layout columns
left, right = st.columns([1, 2])

# Game selection panel
with left:
    st.markdown('<div class="left-box">', unsafe_allow_html=True)
    st.subheader("üîπ Choose a Game")
    for game in games:
        if st.button(game, key=game):
            st.session_state.selected_game = game
    st.markdown('</div>', unsafe_allow_html=True)

# Get selected game
game = st.session_state.get("selected_game", None)

# Game content area
with right:
    if not game:
        st.info("üëà Select a game from the left to start playing.")
    elif game == "Rock Paper Scissors":
        st.header("ü™®üìÑ‚úÇÔ∏è Rock Paper Scissors")
        choices = ["Rock", "Paper", "Scissors"]
        user = st.selectbox("Choose:", choices)
        if st.button("Play RPS"):
            comp = random.choice(choices)
            st.write(f"Computer chose: {comp}")
            if user == comp:
                st.success("It's a tie!")
            elif (user == "Rock" and comp == "Scissors") or \
                 (user == "Paper" and comp == "Rock") or \
                 (user == "Scissors" and comp == "Paper"):
                st.success("You win!")
            else:
                st.error("You lose!")

    elif game == "Guess the Number":
        st.header("üîπ Guess the Number (1‚Äì100)")
        if "target" not in st.session_state:
            st.session_state.target = random.randint(1, 100)
        guess = st.number_input("Enter your guess:", 1, 100)
        if st.button("Check Guess"):
            if guess == st.session_state.target:
                st.success("üéâ Correct!")
                del st.session_state.target
            elif guess < st.session_state.target:
                st.warning("Try higher!")
            else:
                st.warning("Try lower!")

    elif game == "Word Scramble":
        st.header("üå§Ô∏è Word Scramble")
        if "scramble_word" not in st.session_state:
            word = random.choice(["python", "streamlit", "arcade", "banana", "coding"])
            st.session_state.scramble_word = word
            st.session_state.scrambled = ''.join(random.sample(word, len(word)))
        st.write(f"Scrambled word: {st.session_state.scrambled}")
        ans = st.text_input("Your guess:")
        if st.button("Submit Guess"):
            if ans.lower() == st.session_state.scramble_word:
                st.success("Correct! üéØ")
                del st.session_state.scramble_word
                del st.session_state.scrambled
            else:
                st.error("Incorrect!")

    elif game == "Emoji Quiz":
        st.header("üòÑ Emoji Quiz")
        quiz = {
            "üê±üê∂": "animals", "üé¨üçø": "movie", "üéìüìö": "study",
            "üçïüçî": "food", "‚úàÔ∏èüåç": "travel"
        }
        if "emoji_q" not in st.session_state:
            st.session_state.emoji_q = random.choice(list(quiz.items()))
        st.write(f"What do these emojis represent? {st.session_state.emoji_q[0]}")
        guess = st.text_input("Your guess:")
        if st.button("Submit Answer"):
            if guess.lower() == st.session_state.emoji_q[1]:
                st.success("Correct!")
                del st.session_state.emoji_q
            else:
                st.warning("Not quite!")

    elif game == "Score Tracker":
        st.header("üìä Score Tracker")
        st.session_state.setdefault("score", 0)
        col1, col2, col3 = st.columns(3)
        if col1.button("‚ûï"):
            st.session_state.score += 1
        if col2.button("‚ûñ"):
            st.session_state.score -= 1
        if col3.button("üîÑ"):
            st.session_state.score = 0
        st.metric("Current Score", st.session_state.score)

    elif game == "Dice Roller":
        st.header("üé≤ Dice Roller")
        if st.button("Roll Dice"):
            roll = random.randint(1, 6)
            st.success(f"You rolled a {roll} üé≤")

    elif game == "Coin Toss":
        st.header("ü™ô Coin Toss")
        if st.button("Toss Coin"):
            result = random.choice(["Heads", "Tails"])
            st.success(f"It's {result}!")

    elif game == "Math Quiz":
        st.header("‚ûï Math Quiz")
        if "a" not in st.session_state:
            st.session_state.a = random.randint(1, 50)
            st.session_state.b = random.randint(1, 50)
        st.write(f"{st.session_state.a} + {st.session_state.b} = ?")
        ans = st.number_input("Your answer:", step=1)
        if st.button("Submit Answer"):
            if ans == st.session_state.a + st.session_state.b:
                st.success("‚úÖ Correct!")
                del st.session_state.a
                del st.session_state.b
            else:
                st.error("‚ùå Incorrect!")

    elif game == "Hangman":
        st.header("ü™ì Hangman (Lite)")
        if "hangman_word" not in st.session_state:
            st.session_state.hangman_word = random.choice(["python", "apple", "banana"])
            st.session_state.hangman_guesses = []
        display = " ".join([c if c in st.session_state.hangman_guesses else "_" for c in st.session_state.hangman_word])
        st.write(display)
        guess = st.text_input("Guess a letter:").lower()
        if st.button("Submit Letter") and guess:
            if guess in st.session_state.hangman_word:
                st.session_state.hangman_guesses.append(guess)
                st.success("Correct!")
            else:
                st.error("Incorrect.")
        if all(c in st.session_state.hangman_guesses for c in st.session_state.hangman_word):
            st.success("You guessed it! üéâ")
            del st.session_state.hangman_word
            del st.session_state.hangman_guesses

    elif game == "Color Guess":
        st.header("üé® Color Guess")
        colors = ["red", "blue", "green", "yellow", "purple"]
        if "color_target" not in st.session_state:
            st.session_state.color_target = random.choice(colors)
        guess = st.selectbox("Pick a color:", colors)
        if st.button("Guess Color"):
            if guess == st.session_state.color_target:
                st.success("üéØ Correct!")
                del st.session_state.color_target
            else:
                st.warning(f"‚ùå Wrong! It was {st.session_state.color_target}")
                del st.session_state.color_target

'''

with open("app.py", "w") as f:
    f.write(app_code.strip())

print("‚úÖ app.py created.")
```
